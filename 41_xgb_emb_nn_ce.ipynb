# XGBoost, with NAICS embeddings from neural network, plus count encoded NAICS
Retry an XGBoost model but use the NAICS embeddings generated by the standard Neural Network (script 12).  Include the count encoded NAICS feature to see if that helps with unseen or low volume NAICS

*This script takes about 5 minutes on my MacBook Air*

```python
import pandas as pd
import numpy as np
import xgboost as xgb
import warnings
import matplotlib.pyplot as plt
import pickle
```

```python
from pathlib import Path
```

```python
import sklearn as sk
from sklearn.model_selection import RandomizedSearchCV
from sklearn.inspection import permutation_importance, partial_dependence
from scikitplot.metrics import plot_lift_curve
```

```python
from sba_gnn.sba_gnn import sg_plot 
```

## Import Processed Datasets

```python
sba_loans = pd.read_parquet(Path(setup.temp_path).joinpath('01_DATA_transformed.parquet'))
```

```python
embed_df = pd.read_parquet(Path(setup.temp_path).joinpath('12_DATA_embeddings.parquet')) 
```

```python
sba_loans = sba_loans.merge(embed_df.drop(columns='NAICS').rename(columns={'NAICS_orig':'NAICS'}),
                            on='NAICS', how='left')
```

```python
train_df = sba_loans[sba_loans['dset'] == 'train'].copy()
test_df = sba_loans[sba_loans['dset'] == 'test']
val_df = sba_loans[sba_loans['dset'] == 'val']
```

```python
train_df.shape
```

```python
train_df.drop_duplicates('LoanNr_ChkDgt').shape
```

#### Set the predictor features.  Base + embeddings + count encoded NAICS

```python
predictor_features = setup.predictor_features + [c for c in embed_df.columns if c.startswith('emb')] + ['cenc_NAICS']
print(predictor_features)
```

```python
train_df[predictor_features].describe()
```

```python
tcorr = train_df[predictor_features + ['target']].corr()
tcorr.to_csv(Path(setup.temp_path).joinpath('41_REPORT_correlations.csv'))
tcorr[tcorr.index == 'target']
```

```python
with open(Path(setup.temp_path).joinpath('41_DATA_features_predict.pkl'), 'wb') as fout:
    pickle.dump(predictor_features, fout)
```

## Model Fit

```python
best_params_df = pd.read_csv(Path(setup.temp_path).joinpath('03_REPORT_fit_parameter_selected.csv'))
best_params = best_params_df.to_dict(orient='records')[0]
best_params
```

##### Fit using pre-determined parameters

```python
xgb_params = dict(best_params, 
                   **{'objective':"binary:logistic", 'n_estimators':setup.xgb_n_estimators})
```

```python
xgb_model = xgb.XGBClassifier(objective="binary:logistic",
                             n_estimators = setup.xgb_n_estimators)
```

```python
xgb_model = xgb_model.set_params(**xgb_params)
```

```python
xgb_model.fit(train_df[predictor_features], train_df['target'],
              eval_set = [(train_df[predictor_features], train_df['target']),
                         (val_df[predictor_features], val_df['target'])])
```

##### Save the model

```python
xgb_model.save_model(Path(setup.temp_path).joinpath('41_MODEL_xgboost.json'))
```

```python
with open(Path(setup.temp_path).joinpath('41_MODEL_xgboost.pkl'), 'wb') as po:
    pickle.dump(xgb_model, po)
```

## Probability Predictions

```python
pred_df = pd.concat([sba_loans[['LoanNr_ChkDgt', 'dset', 'target', 'dset_naics_holdout']].reset_index(),
                     pd.DataFrame({'predict_prob':
                                   xgb_model.predict_proba(sba_loans[predictor_features])[:,1]})],
                    axis=1) \
    .set_index('index')
```

```python
pred_df.groupby('target')['predict_prob'].agg(['mean', 'std', 'min', 'max', 'median'])
```

## Decision Threshold
Max f1, on training data

```python
pred_train_df = pred_df[pred_df['dset'] == 'train']
```

```python
thresh_tune_data = sg_plot.get_f1_frame(pred_train_df['target'], 
                                        pred_train_df['predict_prob'])
```

```python
thresh_tune_data.to_csv(Path(setup.temp_path).joinpath('41_REPORT_thresh_tune.csv'), index=False)
```

```python
thresh_head = thresh_tune_data.sort_values('f1', ascending=False).head(2)
best_thresh = thresh_head['thresh'].iloc[0]
print(best_thresh)
thresh_head
```

##### Apply decision threshold
All datasets

```python
pred_df['predict_bin'] = np.where(pred_df['predict_prob'] >= best_thresh, 1, 0)
```

```python
pred_df['predict_bin'].value_counts()
```

```python
pred_df.to_parquet(Path(setup.temp_path).joinpath('41_DATA_combined_predictions.parquet'))
```

## PDP Plot - Embeddings

```python
def get_pdp(feature):
    results = partial_dependence(xgb_model, train_df[predictor_features], [feature])
    return pd.DataFrame({'grid_values':results["grid_values"][0],
                       'pdp':results["average"][0]})
```

```python
emb_feat = [c for c in embed_df.columns if c.startswith('emb')]
pdp_res = pd.concat([get_pdp(c) for c in emb_feat], axis=0, keys=emb_feat) 
```

```python
pdp_res = pdp_res.reset_index(level=0) \
    .rename(columns={'level_0':'feature'})
pdp_res.head()
```

```python
nrow = int(np.ceil(len(emb_feat) / 2))
ncol = 2 if len(emb_feat) % 2 == 0 else 3
```

```python
fig, ax = plt.subplots(nrow, ncol, figsize = (10, 8), sharex = False, sharey=True)
for i in range(len(emb_feat)):
    this_row = int(np.floor(i / 2))
    this_col = int(i % 2)
    this_pdp = pdp_res[pdp_res['feature'] == emb_feat[i]]
    #print(f'i:{i}, row:{this_row}, col:{this_col}, feat:{emb_feat[i]}')
    this_pdp.plot(x='grid_values', y='pdp', label=emb_feat[i], ax=ax[this_row, this_col])
    #ax[this_row, this_col].set_title(emb_feat[i])
    #ax[this_row, this_col].legend(loc='upper left')
    ax[this_row, this_col].set_xlabel(None)
plt.tight_layout()
plt.show()
fig.savefig(Path(setup.temp_path).joinpath('41_PLOT_pdp_naics_embeddings.png'),
               bbox_inches='tight')
```
